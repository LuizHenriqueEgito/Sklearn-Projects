# Create Datasets

```python
from sklearn import datasets
import matplotlib.pyplot as plt
```

### Make Moons

Creates two `parabolas` with negative and positive concavity respectively

```python
X, y = datasets.make_moons(
    n_samples=1000,  # Numero de amostras (linhas)
    shuffle=True,  # Se deve embaralhar as amostras
    noise=0.15,  # ruido adicionado aos dados, quanto maior mais misturados eles ficam
    random_state=42,  # escolhe a semente de aleatoriedade
)
```

```python
fig, ax = plt.subplots()
ax.scatter(x=X[:, 0], y=X[:, 1], c=y)
ax.grid(True, linestyle='dotted', c='grey', alpha=0.2)
ax.set_title('Exemplo de dados com make moons')
plt.show()
```

### Make Blobs

Used for `clustering`.

### Make Circles

Creates a `circle` inside another `circle`.

```python
X, y = datasets.make_circles(
    n_samples=1000,
    shuffle=True ,
    noise=None,
    random_state=42 ,
    factor=0.5  # o quanto o circulo de dentro  do de fora
)
```

```python
fig, ax = plt.subplots()
ax.scatter(x=X[:, 0], y=X[:, 1], c=y)
ax.grid(True, linestyle='dotted', c='grey', alpha=0.2)
ax.set_title('Exemplo de dados com make circles')
plt.show()
```

```python
X, y = datasets.make_classification(
    n_samples=1000,
    n_features=2,
    n_informative=2,
    n_redundant=0,
    n_classes=2,
    n_clusters_per_class=2,
    flip_y=0.01,
    class_sep=1.5
)
```

### Make S curve

```python
X, t = datasets.make_s_curve(n_samples=1000)
```

```python
fig, axs = plt.subplots(1,3, figsize=(10,5))
axs[0].scatter(x=X[:, 0], y=X[:, 1], c=t)
axs[1].scatter(x=X[:, 1], y=X[:, 2], c=t)
axs[2].scatter(x=X[:, 0], y=X[:, 2], c=t)

axs[0].set_xlabel('X0')
axs[0].set_ylabel('X1')

axs[1].set_xlabel('X1')
axs[1].set_ylabel('X2')

axs[2].set_xlabel('X0')
axs[2].set_ylabel('X2')

axs[0].grid(True, linestyle='dotted', c='grey', alpha=0.2)
axs[1].grid(True, linestyle='dotted', c='grey', alpha=0.2)
axs[2].grid(True, linestyle='dotted', c='grey', alpha=0.2)
fig.tight_layout()
plt.show()
```
