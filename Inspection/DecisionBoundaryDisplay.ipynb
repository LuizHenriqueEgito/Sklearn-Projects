```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.inspection import DecisionBoundaryDisplay

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier, ExtraTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.svm import SVC, LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
```

```python
models = [
    LogisticRegression(),
    DecisionTreeClassifier(max_depth=5),
    ExtraTreeClassifier(max_depth=5),
    RandomForestClassifier(n_estimators=20, max_depth=5),
    AdaBoostClassifier(n_estimators=20),
    SVC(),
    LinearSVC(),
    KNeighborsClassifier(n_neighbors=5),
    MLPClassifier(hidden_layer_sizes=(2,2), max_iter=10_000),
    LinearDiscriminantAnalysis(),
    QuadraticDiscriminantAnalysis()
]
```

```python
# create a dataset
X, y = make_moons(
    n_samples=1000,  # Numero de amostras (linhas)
    shuffle=True,  # Se deve embaralhar as amostras
    noise=0.15,  # ruido adicionado aos dados, quanto maior mais misturados eles ficam
    random_state=42,  # escolhe a semente de aleatoriedade
)
```

```python
for model in models:
    model.fit(X, y)
    disp = DecisionBoundaryDisplay.from_estimator(
        model, X, response_method="predict",
        xlabel='X0', ylabel='X1',
        alpha=0.5)
    disp.ax_.set_title(type(model).__name__)
    disp.ax_.scatter(X[:, 0], X[:, 1], c=y, edgecolor="k")
    plt.show()
```
